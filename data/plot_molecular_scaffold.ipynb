```python
from rdkit import Chem
from rdkit.Chem.Scaffolds import MurckoScaffold
from collections import Counter
import pandas as pd

from IPython.display import Image
```

```python
bitter = pd.read_csv('BST.csv',encoding = 'gb2312')['SMILES'][0:1946].tolist()
sweet = pd.read_csv('BST.csv',encoding = 'gb2312')['SMILES'][1946:8045].tolist()
tasteless = pd.read_csv('BST.csv',encoding = 'gb2312')['SMILES'][8045:].tolist()
```

# bitter_carbonscaffold

```python
# 提取碳骨架
Bitter_MurckoScaffold = []
for smiles in bitter:
    mol = Chem.MolFromSmiles(smiles)
    scaffold = MurckoScaffold.MurckoScaffoldSmiles(mol=mol)
    Bitter_MurckoScaffold .append(scaffold)
    
Bitter_MurckoScaffold_counts = Counter(Bitter_MurckoScaffold)

sorted_Bitter_MurckoScaffold_counts = sorted(Bitter_MurckoScaffold_counts.items(), key=lambda x: x[1], reverse=True)

sorted_Bitter_MurckoScaffold_counts
```

```python
len(sorted_Bitter_MurckoScaffold_counts)
```

```python
# 输出smiles
for skeleton, count in sorted_Bitter_MurckoScaffold_counts[0:151]:
    print(f'{skeleton}')
```
