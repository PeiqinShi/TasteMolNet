# 删除含有金属元素的SMILES

```python
# 定义函数以检查Smiles是否含有金属元素
def contains_metal(smiles):
    mol = Chem.MolFromSmiles(smiles)
    if mol:
        for atom in mol.GetAtoms():
            if atom.GetAtomicNum() in [3, 11, 13, 19, 20, 25, 26, 29, 30, 42, 47, 50, 73, 74, 78, 79, 80, 82, 83]:
                return True
    return False
```
