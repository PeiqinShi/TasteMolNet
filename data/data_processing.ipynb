```python
import pandas as pd
from rdkit import Chem
from rdkit.Chem import rdMolDescriptors
```

```python
data = pd.read_csv('../data/BST.csv',encoding = 'gb2312')
```

```python
data
```

# 删除含有金属元素的SMILES

```python
# 定义函数以检查Smiles是否含有金属元素
def contains_metal(smiles):
    mol = Chem.MolFromSmiles(smiles)
    if mol:
        for atom in mol.GetAtoms():
            if atom.GetAtomicNum() in [3, 11, 13, 19, 20, 25, 26, 29, 30, 42, 47, 50, 73, 74, 78, 79, 80, 82, 83]:
                return True
    return False
```

```python
# 过滤含有金属元素的SMILES项
filtered_df = data[~data['SMILES'].apply(contains_metal)]
```

```python
filtered_df
```

```python
删除分子量小于30和大于1000的化合物
```

```python
# 添加分子量列
filtered_df['molecular_weight'] = filtered_df['SMILES'].apply(lambda x: rdMolDescriptors.CalcExactMolWt(Chem.MolFromSmiles(x)) if Chem.MolFromSmiles(x) else None)
```

```python
filtered_df
```

```python
# 过滤分子量在20和1600之间的项
filtered_df_2 = filtered_df[(filtered_df['molecular_weight'] >= 25) & (filtered_df['molecular_weight'] <= 1600)]
```

```python
filtered_df_2.to_csv('BST1.csv')
```
